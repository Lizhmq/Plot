```python
def extract(file, regex=r"steps: (\d+)  ppl: (\d+.\d+)"):
    xs, ys = [], []
    with open(file, "r") as f:
        for line in f.readlines():
            lst = re.findall(r"steps: (\d+)  ppl: (\d+.\d+)", line)
            for pair in lst:
                xs.append(int(pair[0]))
                ys.append(float(pair[1]))
    return xs, ys
```
